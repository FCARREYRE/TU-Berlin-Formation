```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Comparaison des temps de trajet avec et sans la no-car zone

df_original = pd.read_csv('./output_original_run/travelTimeByPerson.txt', names=['VehicleId','TravelTime'], sep=';')
df_modified = pd.read_csv('./output_dm1/travelTimeByPerson.txt' , names=['VehicleId','TravelTime'], sep=';')


table = [df_original['VehicleId'].tolist(), df_original['TravelTime'].tolist(), df_modified['TravelTime'].tolist()]


x = [i for i in range (len(table[0]))]


fig = plt.figure(figsize = (20,6))
     
y1 = table[1]
y1 = [a/2*24/30 for a in y1]
y2 = table[2]
y2 = [a/2*24/30 for a in y2]
y3 = [np.mean(y1) for i in range(len(y1))]
y4 = [np.mean(y2) for i in range(len(y2))]
print(np.mean(y1))
print(np.mean(y2))

plt.bar(x,y2, label = 'With the no-car zone', color = 'blue' )
plt.bar(x,y1, label='Without the no-car zone', color = 'orange')
plt.plot(x, y3, '--', label = 'Mean time without the no car zone',color = 'orange')
plt.plot(x, y4, '--', label = 'Mean time with the no car zone',color = 'blue')

plt.xlabel('Person ID')
plt.ylabel('Daily travel time')
plt.title('Comparison of the travel time with and without the no-car zone')
plt.legend()
plt.show()
```

```python
df_altlinks_original = pd.read_csv('./output_original_run/number_of_entries.txt', names=['0','1','2','3','4','5','6','7'], sep=';')
df_altlinks_modified = pd.read_csv('./output_dm1/number_of_entries.txt' , names=['0','1','2','3','4','5','6','7'], sep=';')

# print(df_altlinks_original)
table_original = [df_altlinks_original[str(i)].tolist()[:31] for i in range(0,8)]
table_modified = [df_altlinks_modified[str(j)].tolist()[:31] for j in range(0,8)]

fig = plt.figure(figsize = (20,25)) 


for k in range(1,9):
    plt.subplot(4,2,k)
    plt.plot(table_original[k-1], label = 'Without no-car zone', color='blue')
    plt.plot(table_modified[k-1], label = 'With no-car zone', color = 'orange')
    plt.plot([np.mean(table_modified[k-1]) for i in range(31)], '--', color = 'orange', label = 'Mean with no car zone')
    plt.plot([np.mean(table_original[k-1]) for i in range(31)], '--', color = 'blue', label = 'Mean without no car zone')
    plt.xlabel('Time')
    plt.ylabel('Number of cars')
    plt.legend()







fig.suptitle('Comparison of traffic on alternative routes')
plt.legend()
plt.show()
```
